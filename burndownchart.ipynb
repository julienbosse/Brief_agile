```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
import numpy as np

data = pd.DataFrame()

data["y1"]= [171-34.2*i for i in range(0,6)]
data["x1"]=[1,2,3,4,5,6]

data["y2"]=[171,161,140,0,0,0]
data["x2"]=[1,2,3,4,5,6]

sns.set_style("white")

fig, ax = plt.subplots(figsize=[10,6])
ax.plot(data["x1"], data["y1"],"--")
ax.plot(data["x2"], data["y2"])
ax.grid()

labels_x=["0","Mar 15","Mer 16","Jeu 17","Ven 18","Lun 21","Mar 22"]
ax.set_xticklabels(labels_x, rotation=0)

# .despine permet de supprimer les axes de haut et de droite
# top=True, right=True, left=False, bottom=False

sns.despine()

# paper, notebook, talk et poser
sns.set_context("notebook")
plt.show()
plt.close()
```
