```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("donnees/housing.csv")
```

```python
data
```

```python
data_columns = data.columns
data_columns
```

```python
nbdeNAN_total_bedrooms = data[data_columns][data[data_columns[4]].isna()==True].shape[0]
nbdeNAN_total_bedrooms
```

```python
donnees_boxplot=data[['median_house_value','total_rooms','population','total_bedrooms','households','median_income']]
```

```python
plt.figure(figsize=[9,6])
sns.set_context("notebook",rc={'xtick.labelsize':18,'ytick.labelsize':13})
palette=sns.color_palette("YlOrBr")
sns.set_palette(palette)
plt.subplot(111)

sns.boxplot(data=donnees_boxplot,dodge=True)
plt.xticks(rotation=90)
plt.show()
plt.close()
```
