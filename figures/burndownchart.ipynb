```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
import numpy as np

data = pd.DataFrame()

data["y1"]= [90-22.5*i for i in range(0,5)]
data["x1"]=[1,2,3,4,5]

data["y2"]=[90,57,38,18,0]
data["x2"]=[1,2,3,4,5]

sns.set_style("white")

fig, ax = plt.subplots(figsize=[12,8])
ax.plot(data["x1"], data["y1"],"--")

ax.plot(data["x2"], data["y2"],"-o")

ax.grid()

labels_x=["0","Mer 23","","Jeu 24","","Ven 25","","Lun 28","","Mar 29"]
ax.set_xticklabels(labels_x, rotation=0)

# .despine permet de supprimer les axes de haut et de droite
# top=True, right=True, left=False, bottom=False

sns.despine()

# paper, notebook, talk et poser
sns.set_context("notebook")
plt.show()
# fig.savefig("figures/burndown_sprint1")
plt.close()
```
